## Technical Stats Of The Dataset

```python
import pandas as pd
import matplotlib as plt

file_path = 'sales.csv'
df = pd.read_csv(file_path)
print(df.describe())
```

### Questions
### 1. Give top companies having high sales.

```python
print(df[df['Sales_Qrt_Category']=="High Sales"])
```

### 2. Give companies having low sales and small capital.

```python
print(df[
          (df.Sales_Qrt_Category == 'Low Sales') &
          (df.Market_Cap_Category == "Small Cap")
        ])
```

### 3. Give bottom 5 companies having very low sales and very small capital.

```python
print(df[
          (df.Sales_Qrt_Category == 'Very Low Sales') &
          (df.Market_Cap_Category == 'Very Small Cap')
        ].tail())
```

### 4. Give the total sales of all companies in crores and total mean of all companies' capitals in crores.

```python
print(df.agg({'Sales Qtr - Crore' : 'sum', 'Mar Cap - Crore' : 'mean'}))
```

### Data Visualization
### Line plot of sales in crores

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'sales.csv'
df = pd.read_csv(file_path)
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Sales Qtr - Crore'], marker='o', linestyle='-', color='b', label='Sales (in crores)')
plt.xlabel('Companies')
plt.ylabel('Sales in crores')
plt.title('Sales of companies in crores')
plt.grid(True)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
